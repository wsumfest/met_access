```python
"""
I noticed a significant number of NaN columns in the dataframe.
I want to train a classifier for the Medium column. Therefore, I will drop all values with NaN in their Medium column.
"""
# dataframe_ = dataframe[pandas.notnull(dataframe["Medium"])]

dataframe.dropna(subset=["Medium"], inplace=True)

print "Dropped {0} rows with empty Medium field.".format(intial_num_rows - len(dataframe.index))
```

```python
"""
Another very useful field seems to be classification. Upon exploration, there appears to be a significant number of rows where
this value is empty. Moreover, these rows map to a large number of Mediums. This does not give me any insight.
"""
print len(dataframe[dataframe["Classification"].isnull()])

print len(set(dataframe[dataframe["Classification"].isnull()]["Medium"]))
```

```python
"""
I want to see the columns with the most/least significant number of NaN values.
"""

for _ in column_names:
    print "Column: {0}, Nan Values: {1}".format(_, len(dataframe[dataframe[_].isnull()]))
```

```python
"""
Given this information above, it appears all columns can be useful in training our classifier. 
It will now be imperative to partition the data for my classifier.
Upon doing this exercise, I noticed special characters, most likely human error, creating different items. 
I will clean it up by stripping special chars.
"""
import re

def strip_chars(vals):
    if type(vals) != str:
        return vals
    return re.sub('[^A-Za-z0-9\s]+', '', vals).strip().replace("\r", "").replace("\n", "")


dataframe["Medium"] = dataframe["Medium"].apply(strip_chars)

# unique_mediums = list(set(dataframe["Medium"]))
# transformed = set()
# for item in unique_mediums:
#     transformed.add(strip_chars(item))
    
# print len(unique_mediums)
# print len(transformed)

# _map = dict()

# for medium in unique_mediums:
# #     print "Medium: {0}, #Rows: {1}".format(medium, len(dataframe.loc[dataframe["Medium"] == medium]))
#     _map[medium] = len(dataframe.loc[dataframe["Medium"] == medium])
dataframe.groupby("Medium").count().sort_values(by=["Object Number"], ascending=False)
```
