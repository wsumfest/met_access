```python
import pandas
import numpy
import sklearn.tree as tr
import sklearn.neighbors as nb
```

```python
SOURCE_DATA = "/development/met_access/data/raw_met_data.csv"
```

```python
dataframe = pandas.read_csv(SOURCE_DATA)
```

```python
#uncomment to print out top rows to explore data.
dataframe.head(10)
```

```python
column_names = dataframe.columns
print(column_names)

dataframe.dtypes

for col in column_names:
    print('{0} is unique: {1}'.format(col, dataframe[col].is_unique))
    
# print dataframe.index
intial_num_rows = len(dataframe.index)
    
"""
I explored the column names, datatypes, and index. As well as uniquely identifiable columns. 
I dropped all unique identifiers as they do not give insight into the data.
"""
dataframe = dataframe.drop(["Link Resource"], axis=1)
column_names = dataframe.columns
```

```python
"""
I noticed a significant number of NaN columns in the dataframe.
I want to train a classifier for the Medium column. Therefore, I will drop all values with NaN in their Medium column.
"""
# dataframe_ = dataframe[pandas.notnull(dataframe["Medium"])]

dataframe.dropna(subset=["Medium"], inplace=True)

print("Dropped {0} rows with empty Medium field.".format(intial_num_rows - len(dataframe.index)))
```

```python
"""
Another very useful field seems to be classification. Upon exploration, there appears to be a significant number of rows where
this value is empty. Moreover, these rows map to a large number of Mediums. This does not give me any insight.
"""
print(len(dataframe[dataframe["Classification"].isnull()]))

print(len(set(dataframe[dataframe["Classification"].isnull()]["Medium"])))
```

```python
"""
I want to see the columns with the most/least significant number of NaN values.
"""

for _ in column_names:
    print("Column: {0}, Non-Nan Values: {1}".format(_, len(dataframe[dataframe[_].notnull()])))
```

```python
"""
I noticed several fields with only one value. I will drop all of these as they are not useful.
"""
counts = dataframe["Medium"].value_counts()
indices = counts[counts > 1].index
dataframe = dataframe[dataframe["Medium"].isin(indices)]
```

```python
# print(dataframe[dataframe["State"].notnull()]["Medium"].value_counts())

"""
The state field does not seem to be particularly valuable. It can be dropped.
"""

# print(dataframe[dataframe["River"].notnull()]["Medium"].value_counts())

"""
The River field does not seem to be particularly valuable. It can be dropped.
"""

# print(dataframe[dataframe["Locus"].notnull()]["Medium"].value_counts())

"""
The locus field seems useful. It will be kept.
"""
# print(dataframe[dataframe["Locale"].notnull()]["Medium"].value_counts())

"""
The locale field seems useful. It will be kept.
"""
dataframe = dataframe.drop(["River", "State"], axis=1)
column_names = dataframe.columns

```

```python
"""
Given this information above, it appears all columns can be useful in training our classifier. 
It will now be imperative to partition the data for my classifier.
Upon doing this exercise, I noticed special characters, most likely human error, creating different items. 
I will clean it up by stripping special chars.
"""
import re

def strip_chars(vals):
    if type(vals) != str:
        return vals
    return re.sub('[^A-Za-z0-9\s]+', '', vals).strip().replace("\r", "").replace("\n", "")


dataframe["Medium"] = dataframe["Medium"].apply(strip_chars)

# unique_mediums = list(set(dataframe["Medium"]))
# transformed = set()
# for item in unique_mediums:
#     transformed.add(strip_chars(item))
    
# print len(unique_mediums)
# print len(transformed)

# _map = dict()

# for medium in unique_mediums:
# #     print "Medium: {0}, #Rows: {1}".format(medium, len(dataframe.loc[dataframe["Medium"] == medium]))
#     _map[medium] = len(dataframe.loc[dataframe["Medium"] == medium])
dataframe.groupby("Medium").count().sort_values(by=["Object Number"], ascending=False)
```

```python
"""
Time to prepare and partition data.

First, we factorize our mediums so we can display the output.

"""

values, indices = pandas.factorize(dataframe["Medium"])


prepared_dataframe = dataframe.apply(lambda x: pandas.factorize(x)[0])

N = len(dataframe.index)

TRAINING_DATA_SIZE = 1/3.0 * N
TEST_DATA_SIZE = 1/3.0 * N
RECLASSIFY_DATA_SIZE = 1/3.0 * N

```

```python
import math

training_dataframe = prepared_dataframe.sample(int(math.floor(TRAINING_DATA_SIZE)))
```

```python
X = training_dataframe.drop(['Medium'], axis=1)
Y = training_dataframe["Medium"]

model = nb.KNeighborsClassifier()
```

```python
model.fit(X, Y)
```

```python
model.score(X, Y)
```

```python
testing_data = prepared_dataframe

X = testing_data.drop(['Medium'], axis=1)
Y = testing_data["Medium"]

model.score(X, Y)
```
